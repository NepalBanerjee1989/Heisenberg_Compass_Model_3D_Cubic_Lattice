```python
############################################################
# Plotting the data from MC calculation 3D_XYZ ISOTROPIC   #
############################################################
import numpy as np
import matplotlib.pyplot as plt
data1=np.loadtxt('ISO_9')
data2=np.loadtxt('ISO_7')
data3=np.loadtxt('ISO_5')
plt.figure(figsize=(14,10))
font1 = {'family':'serif','color':'blue','size':20}

######### Magnetization vs T ###############################
plt.subplot(221)
#plt.title(' M vs T ')
plt.plot(data1[:,0], data1[:,1],  'bo--',label='9x9x9' ,lw=1)
plt.plot(data2[:,0], data2[:,1],  'ro--',label='7x7X7',lw=1)
plt.plot(data3[:,0], data3[:,1],  'ko--',label='5x5X5',lw=1)
plt.xlabel("Average Pulse", fontdict = font1,fontsize=20)
plt.ylabel("Calorie Burnage", fontdict = font1,fontsize=20)
plt.axvline(1.1,linewidth=3.0, color='k',linestyle='--')
T=data1[:,0]
M=data1[:,1]
plt.text(0.3,0.5, 'Ferro', fontsize=30, rotation=0)
plt.text(2.0,0.5, 'Para', fontsize=30, rotation=0)
plt.fill_between(T,0.0,.80, where=(T <= 1.1), facecolor='blue', alpha=0.40)
plt.fill_between(T,0.0,0.80, where=(T >=1.1), facecolor='red',  alpha=0.40)
#plt.text(-0.70,1,'a)',fontsize=20, rotation=0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.xlabel('T')
plt.ylabel('M')


###########################################################
plt.subplot(222)
#plt.title('$\chi$ vs T ')
plt.plot(data1[:,0],data1[:,2],'bo--',label='9x9x9',lw=1)
plt.plot(data2[:,0],data2[:,2],'ro--',label='7x7x7',lw=1)
plt.plot(data3[:,0],data3[:,2],'ko--',label='5x5x5',lw=1)
plt.axvline(1.1,linewidth=3.0, color='k',linestyle='--')
T=data1[:,0]
M=data1[:,1]
plt.text(0.3,2.0, 'Ferro', fontsize=30, rotation=0)
plt.text(2.0,2.0, 'Para', fontsize=30, rotation=0)
plt.fill_between(T,0.0,3.0, where=(T <= 1.1), facecolor='blue', alpha=0.40)
plt.fill_between(T,0.0,3.0, where=(T >= 1.1), facecolor='red',  alpha=0.40)
#plt.text(-0.60,4,'b)',fontsize=20, rotation=0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.xlabel("Average Pulse", fontdict = font1,fontsize=20)
plt.ylabel("Calorie Burnage", fontdict = font1,fontsize=20)
plt.xlabel('T')
plt.ylabel('$\chi$')

############################################################
plt.subplot(223)
#plt.title('$ E $ vs T ')
plt.plot(data1[:,0], data1[:,3],  'bo--',label='9x9x9' ,lw=1)
plt.plot(data2[:,0], data2[:,3],  'ro--',label='7x7x7',lw=1)
plt.plot(data3[:,0],data3[:,3],'ko--',   label='5x5x5',lw=1)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend()
plt.xlabel("Average Pulse", fontdict = font1,fontsize=20)
plt.ylabel("Calorie Burnage", fontdict = font1,fontsize=20)
plt.axvline(1.1,linewidth=3.0, color='k',linestyle='--')
T=data1[:,0]
M=data1[:,1]
plt.text(0.3,-1.5, 'Ferro', fontsize=30, rotation=0)
plt.text(2.0,-1.5, 'Para', fontsize=30, rotation=0)
plt.fill_between(T,-4.0,0.0, where=(T <= 1.1), facecolor='blue', alpha=0.40)
plt.fill_between(T,-4.0,0.0, where=(T >= 1.1), facecolor='red', alpha=0.40)
#plt.text(-0.70,0,'c)',fontsize=20, rotation=0)
plt.xlabel('T')
plt.ylabel('E')

#############################################################
plt.subplot(224)
#plt.title('$ Cv $ vs T ')
plt.plot(data1[:28,0], data1[:28,4],  'bo--',label='9x9x9' ,lw=1)
plt.plot(data2[:29,0], data2[:29,4],  'ro--',label='7x7x7',lw=1)
plt.plot(data3[:29,0], data3[:29,4],  'ko--',label='5x5x5',lw=1)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Average Pulse", fontdict = font1,fontsize=20)
plt.ylabel("Calorie Burnage", fontdict = font1,fontsize=20)
plt.axvline(1.1,linewidth=3.0, color='k',linestyle='--')
T=data1[:,0]
M=data1[:,1]
plt.text(0.3,1.55, 'Ferro',fontsize=30, rotation=0)
plt.text(2.0,1.55, 'Para', fontsize=30, rotation=0)
plt.fill_between(T,0.0,1.75, where=(T<=1.1), facecolor='blue', alpha=0.40)
plt.fill_between(T,0.0,1.75, where=(T>=1.1), facecolor='red',  alpha=0.40)
#plt.text(-0.60,2.5,'d)',fontsize=20, rotation=0)
plt.xlabel('T')
plt.ylabel('$C_V$')
plt.legend()
plt.show
plt.savefig('ISO_XYZ_p_0.6.png')
```
