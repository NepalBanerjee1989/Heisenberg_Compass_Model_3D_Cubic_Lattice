```python
#########################################
# Plotting the data from MC calculation #
#########################################
import numpy as np
import matplotlib.pyplot as plt
data1 =np.loadtxt('HSO_6')
data2 =np.loadtxt('HSO_8')
data3 =np.loadtxt('HSO_10')
#data4 =np.loadtxt('HSO_test_NEW_30')
plt.figure(20)
#########################################
######### Magnetization vs T ############
#########################################
#plt.subplot()
#plt.title('M vs T ')
#plt.plot(data1[:,0], data1[:,1], 'b--',label='5 x 5')
font1={'family':'serif','color':'blue','size':10}
font2={'family':'serif','color':'darkred','size':10}
plt.plot(data1[:,0],data1[:,1],'ro-',label='6x6x6',lw=2)
plt.plot(data2[:,0],data2[:,1],'go-',label='8x8x8',lw=2)
plt.plot(data3[:,0],data3[:,1],'mo-',label='10x10x10',lw=2)
##########################################
##########################################
plt.xlabel("Average Pulse",fontdict=font1,fontsize=20)
plt.ylabel("Calorie Burnage",fontdict=font1,fontsize=20)
#Set the font size of xticks
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('T')
plt.ylabel('M')
plt.legend() 
plt.show
plt.savefig('HK_J_1_K_1.png',bbox_inches = 'tight')
```
