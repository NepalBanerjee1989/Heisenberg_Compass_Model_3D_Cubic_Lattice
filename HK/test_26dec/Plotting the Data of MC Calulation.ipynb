```python
###################################################################################
# Plot the initial and final Heisenberg Spin Configuration over the unit Sphere   #
###################################################################################
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm ,colors

# creat a Sphere 
r=1
pi=np.pi
cos=np.cos
sin=np.sin
phi,theta =np.mgrid[0.0:pi:100j , 0.0:2.0*pi:100j ]

x=r*sin(phi)*cos(theta)
y=r*sin(phi)*sin(theta)
z=r*cos(phi)


fig=plt.figure()
ax= fig.add_subplot(221, projection='3d')

#import random initial spin components data 
data =np.loadtxt('random_initial_spin_5')
xx=data[:,3]
yy=data[:,4]
zz=data[:,5]

ax.plot_surface(x,y,z,rstride=1,cstride=1,color='w',alpha=0.6,linewidth=0)
ax.scatter(xx,yy,zz,color="m",s=5)

plt.title('Initial random Heisenberg Spin')
ax.set_xlabel('$S_x$')
ax.set_ylabel('$S_y$')
ax.set_zlabel('$S_z$')

fig=plt.figure()

ax= fig.add_subplot(221, projection='3d')

#import final Heisenberg spin components data after transition
#data1 =np.loadtxt('spin_after_transition_5')
#xxx=data1[:,3]
#yyy=data1[:,4]
#zzz=data1[:,5]

#ax.plot_surface(x,y,z,rstride=1,cstride=1,color='w',alpha=0.6,linewidth=0)
#ax.scatter(xxx,yyy,zzz,color="m",s=5)


#plt.title('Heisenberg Spin After Transition')


ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])

ax.set_xlabel('$S_x$')
ax.set_ylabel('$S_y$')
ax.set_zlabel('$S_z$')

ax.set_aspect("equal")
plt.tight_layout()

plt.show()
plt.savefig('SPIN_configuration.png')    
```
